```
# Empirical Cumulative Distributive Functions

def ecdf(data):
  n = len(data)
  x = np.sort(data)
  y = np.arange(1, n+1) / n
  return x, y

x_so2, y_so2 = ecdf(df['SO2'])
x_no2, y_no2 = ecdf(df['NO2'])
x_pm10, y_pm10 = ecdf(df['PM10'])
_ = plt.plot(x_so2, y_so2,  linestyle='--')
_ = plt.plot(x_no2, y_no2,  linestyle='--')
_ = plt.plot(x_pm10, y_pm10,  linestyle='--')

#annotate the plot
plt.legend(('SO2', 'NO2', 'PM10'), loc='lower right')
_ = plt.xlabel('Pollution chemicals')
_ = plt.ylabel('ECDF')
plt.margins(0.05)
plt.show()
```
